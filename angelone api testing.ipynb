```python
import http.client

# Define the connection
conn = http.client.HTTPSConnection("apiconnect.angelone.in")

# Define the payload using f-strings for clarity
client_id = "AAAE715337"  # Replace with your Angel One client ID
client_pin = "2006"  # Replace with your Angel One password
totp_code = "550804"  # Replace with your TOTP code
state_variable = "STATE_VARIABLE"  # Replace with your state variable
api_key = "BUh0Jiog "  # Replace with your actual Angel One API key

payload = f"""
{{
    "clientcode": "{client_id}",
    "password": "{client_pin}",
    "totp": "{totp_code}",
    "state": "{state_variable}"
}}
"""

# Define headers
headers = {
    "Content-Type": "application/json",
    "Accept": "application/json",
    "X-UserType": "USER",
    "X-SourceID": "WEB",
    "X-ClientLocalIP": "CLIENT_LOCAL_IP",  # Replace with your local IP
    "X-ClientPublicIP": "CLIENT_PUBLIC_IP",  # Replace with your public IP
    "X-MACAddress": "MAC_ADDRESS",  # Replace with your MAC address
    "X-PrivateKey": api_key  # Include the API Key
}

# Send the request
conn.request(
    "POST", 
    "/rest/auth/angelbroking/user/v1/loginByPassword", 
    payload, 
    headers
)

# Get and print the response
res = conn.getresponse()
data = res.read()
print(data.decode("utf-8"))
json_data=data.decode("utf-8")
```

```python
import json 
response_json = json.loads(json_data)

# Extract the actual jwtToken value
access_token = response_json["data"]["jwtToken"]
print(access_token)
```
